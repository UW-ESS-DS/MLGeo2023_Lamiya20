$$ Rrup = \sqrt(Rrup'^2+Ry^2)$$

And in our case, assuming the case is 2 (not 8 as site is n footwall), the equation for calculating the Rrup is shown below:
$$ Rrup' = \sqrt(Rx^2+Ztor^2)$$ for $$Rx <Ztor*tan\delta$$

* Since in this case, we are considering the case is 4, so Ry is 0
* Rrup and Rrup' are bothe the same
* Also, for case 4, $$\alpha= 90 deg$$ 
* As a result, Rx = Rjb (sign will be opposite since Rx is in footwall in this case)
From above table, $$ Rx = Rjb*sin\alpha$$:
For $$\alpha= 90 deg$$ ; $$ Rx = Rjb$$

In general, a taller building is more sensitive to longer period motions.


### <span style="color:Teal">▽ Based on the BCHydro model, the spectral acceleration is found to be higher than the target value within the period range of up to 2.5 seconds ( above 25 story), indicating that it is safe within this range including PGA. However, for periods greater than 2.5 seconds, the model provides lower spectral acceleration values, suggesting that caution should be exercised. Caution is needed because earthquakes with longer periods can have higher amplitudes, which can cause greater damage to structures. Additionally, the frequency content of seismic waves varies with different periods, and the spectral acceleration value is a measure of the intensity of ground shaking at a specific period. Therefore, if the spectral acceleration values are lower for longer periods, it implies that the structure may experience lower ground shaking intensities at these frequencies, which could lead to underestimation of the seismic hazard.  </span>

```python
df = pd.read_excel('assignment 2(q4-c.data from peer naga)/Sparks.2011.Sa.1.xlsx', sheet_name='scaled Psa')
df_1 = pd.read_excel('assignment 2(q4-c.data from peer naga)/Sparks.2011.Sa.1.xlsx', sheet_name='Unscaled PEER')
# print(df)
# read a xls file from "q3.0.2 T. Kocaeli earthquake" folder containing "Kocaeli.0.2Sa.xlsx" file and make dataframe,
df_a = pd.read_excel('Norfolk, Virginia.xlsx', sheet_name='Sheet1')
df_b = pd.read_excel(' Wilshire Boulevard.xlsx', sheet_name='Sheet1')
plt.loglog(df.iloc[:, 0], df.iloc[:, 1], label='Matched Sa for Sparks(Luther) 2011(scaled)', linewidth = 3)
plt.loglog(df_a.iloc[:, 0], df_a.iloc[:, 1], label='Norfolk, Virginia Sa', linewidth = 3)
plt.loglog(df_1.iloc[:, 0], df_1.iloc[:, 1], label='Sa for Sparks.2011(Luther)(unscaled)', linewidth = 3)
plt.loglog(df_b.iloc[:, 0], df_b.iloc[:, 1], label='Sa for Sparks( Wilshire 2011(scaled)', linewidth = 3)
plt.xlabel('Period (s)')
plt.ylabel('PSA (g)')
plt.grid(True, which='both')
plt.minorticks_on()
plt.rcParams['axes.facecolor'] = 'white'
plt.legend()
plt.show()
```

### <span style="color:Purple">Scaling factor used for both motion are shown above in the table. For Luther Middle school station, scaling factor used to match the target Sa is 2.97 whereas Wilshire station, factor is 3.39. Period range 0.5 to 2 sec, Luther station matched for the whole range more or less perfectly, but at 0.13 second, predicted Sa is almost 200% than target one. This indicates that there may be some discrepancies or errors in the scaling factor used or the predictive model itself. In period range, like 0.4 second,the Sa of predicted motios is slightly below the target Sa. From hazard perspective, if the predicted Sa values are significantly higher or lower than the actual Sa values, it can lead to underestimation or overestimation of the potential damage and risk associated with an earthquake. 
### <span style="color:Purple">Moreover, inaccurate predictions of ground motion can also lead to incorrect estimates of other seismic parameters, such as peak ground velocity (PGV) and Arias Intensity, which are used to assess the potential hazard of an earthquake.
